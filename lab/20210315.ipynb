```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_weighted_edges_from([
    ("Wien", "Linz", 184.4),
    ("Wien", "Hagenberg", 180),
    ("Wien", "Graz", 200.1),
    ("Wien", "Salzburg",295),
    ("Linz", "Hagenberg",23),
    ("Linz", "Graz",220.9),
    ("Linz", "Salzburg",132.5),
    ("Salzburg", "Steyr",134),
    ("Salzburg", "Graz",296),
    ("Salzburg", "Wels",108),
    ("Graz", "Steyr",191),
    ("Graz", "Wels",196),
    ("Wels", "Steyr",45.2),
    ("Wels", "Hagenberg",57.2),
])

nx.draw(G, with_labels= True)
plt.show()

fh = G.subgraph(["Linz", "Hagenberg", "Wels", "Steyr"])

nx.draw(fh, with_labels= True)

nx.is_connected(G)
```

```python
def depth_search(graph, seen, current, action):
    if current in seen:
        return seen
    seen = seen + [current]
    action(current, seen)

    for n in G[current]:
        depth_search(graph, seen, n, action)
    return seen

def is_connected(graph, seen = None, current = None):
    if seen is None :
        seen = []
    if current is None:
        current = "Linz"

    seen = depth_search(graph, seen, current, lambda c,p: None)

    return len(seen) == graph.order()

print(is_connected(G))
```
